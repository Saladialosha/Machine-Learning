```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn import tree
```

```python
df = pd.read_csv("salaries.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
temp_df = df
temp_df.company = le.fit_transform(df.company) 
temp_df.job = le.fit_transform(df.job)
temp_df.degree = le.fit_transform(df.degree)
temp_df
```

```python
X,Y = temp_df[['company','job','degree']], df['salary_more_then_100k']
```

```python
from sklearn.model_selection import train_test_split
train_X,test_X,train_Y,test_Y = train_test_split(X,Y,test_size=0.2)
```

```python
model = tree.DecisionTreeClassifier()
model.fit(train_X.values,train_Y)
```

```python
model.score(test_X.values,test_Y)
```

```python
model.predict([[2,0,0]])
```
